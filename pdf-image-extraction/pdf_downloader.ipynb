```python
import json
import os
# Lấy dữ liệu metadata
path = '/content/drive/MyDrive/metadata'
report_list = []
for i in os.listdir(path):
  f = open(os.path.join(path, i))
  data = json.load(f)
  report_list+=data
```

```python
import requests
import time
from tqdm import tqdm
# Lưu dữ liệu metadata
save_path = '/content/drive/MyDrive/pdf'
# with open("/content/drive/MyDrive/status.json", "r") as f:
#     status_list = json.load(f)
status_list = []
for report in tqdm(report_list):
  r = requests.get(report['URL'])
  if r.status_code == 200:
    with open(os.path.join(save_path, '{}.pdf'.format(report['TÊN CÔNG TY'].replace('/', ''))), 'wb') as fd:
      fd.write(r.content)
    status_list.append(True)
  else:
    status_list.append(False)
  with open("/content/drive/MyDrive/status.json", "w") as outfile:
    json.dump(status_list, outfile)
  time.sleep(10)
```
